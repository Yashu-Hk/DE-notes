```python
import pandas as pd
```

```python
df=pd.read_csv("friends.csv")
```

```python
df
```

```python
df["height(cm)"]=pd.to_numeric(df["height(cm)"],errors='coerce') #invalid or errors are converted into nan values
```

```python
df["weight(kg)"]=pd.to_numeric(df["weight(kg)"],errors='coerce')
```

```python
df["spend_C"]=pd.to_numeric(df["spend_C"],errors='coerce')
```

```python
df
```

```python
df.isna().sum()
```

```python
df[['age','sex']]= df.age_sex.str.split("_",expand=True)
```

```python
df.drop(columns=["age_sex"],inplace=True)
```

```python
df.head(2)
```

```python
#ht -ffill wt- bfill remaining -fill by mean
df['height(cm)']=df['height(cm)'].fillna(method='ffill')
df['height(cm)'].head()
```

```python
df["weight(kg)"]=df["weight(kg)"].fillna(method='bfill')
```

```python
ageMean=np.mean(df["age"])
ageMean
```

```python
df["age"]=df["age"].fillna(ageMean)
```

```python
import numpy as np
Amean=np.mean(df['spend_A'])
Amean
```

```python
df["spend_A"]=df["spend_A"].fillna(Amean)
```

```python
Bmean=np.mean(df['spend_B'])
Bmean
```

```python
df["spend_B"]=df["spend_B"].fillna(Bmean)
```

```python
Cmean=np.mean(df['spend_C'])
Cmean
```

```python
df["spend_C"]=df["spend_C"].fillna(Cmean)
```
