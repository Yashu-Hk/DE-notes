```python
#PLot sine graph with theta values 0,30,45,60,90
import math
x=[0,30,45,60,90]
y=[math.sin(i) for i in x]
print(y)
```

```python
plt.xlabel("Theta values")
plt.ylabel("Sine values")
plt.title("Sine Graph")
plt.plot(x,y)
plt.show()
```

```python
x=[0,30,45,60,90]
sinx=[math.sin(i) for i in x]
cosx=[math.cos(i) for i in x]
plt.xlabel("Theta values")
plt.ylabel("Sine-Cosine values")
plt.title("Sine-Cosine Graph")
plt.plot(x,sinx,label="sin")
plt.plot(x,cosx,label="cos")
plt.legend(loc="lower left")
plt.show()
```

```python
import csv
fp=open('sampledt.csv','r')
rows=csv.DictReader(fp)
dname=['hr','fin','mark']
colors = ['blue', 'green', 'orange']
hr=[]
mark=[]
fin=[]
for row in rows:
    if row.get('dname')=='hr':
        hr.append(eval(row.get('sal')))
    elif row.get('dname')=='fin':
        fin.append(eval(row.get('sal')))
    elif row.get('dname')=='mark':
        mark.append(eval(row.get('sal')))
# print(hr)
# print(fin)
# print(mark)
x=np.array([1,2,3])
y=[sum(hr),sum(fin),sum(mark)]
my_xsticks=['HR','FINANCE','MARKETING']
plt.xticks(x,dname)
# plt.xticks(x,my_xsticks)
plt.xlabel('Department')
plt.ylabel('Total Salary')
plt.title('Total Salaries by Department (sampledt.csv)')
# plt.legend(loc='upper center')
plt.bar(x,y,color=colors,label=dnames)
```
